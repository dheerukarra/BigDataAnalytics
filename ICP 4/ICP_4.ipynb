```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://drive.google.com/uc?id={}'.format('16b7HOtAZ5w1-ROp_sZHmsdeplbGGSNFI')
df = pd.read_csv(url)

# Show the basic statistical description about the data.
stat_description = df.describe()

print(stat_description)

# Check if the data has null values and replace with the mean.
df.fillna(df.mean(numeric_only=True), inplace=True)

# Aggregate data for 'Duration' and 'Calories'.
agg_data = df[['Duration', 'Calories']].agg(['min', 'max', 'count', 'mean'])

print(agg_data)

# Filter for calories between 500 and 1000.
filtered_calories_500_1000 = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]

print(filtered_calories_500_1000)

# Filter for calories > 500 and pulse < 100.
filtered_calories_pulse = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]

print(filtered_calories_pulse)

# Create a new dataframe without “Maxpulse”.
df_modified = df.drop(columns=['Maxpulse'])

# Convert Calories to int.
df['Calories'] = df['Calories'].astype(int)

# Scatter plot for Duration and Calories.
plt.figure(figsize=(10,6))
plt.scatter(df['Duration'], df['Calories'], color='blue', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.show()
```

```python
import matplotlib.pyplot as plt


languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700', '#C71585']
explode = (0.1, 0, 0, 0, 0, 0)

# Plotting the pie chart
plt.figure(figsize=(10, 7))
plt.pie(popularity, explode=explode, labels=languages, colors=colors, autopct='%1.1f%%', shadow=False, startangle=140)
plt.title("Popularity of Programming Languages")
plt.show()

```
